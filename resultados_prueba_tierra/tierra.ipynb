```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Cargamos dataframe

```
df = pd.read_csv('/work/Tierra calculos.csv')
df
```

### Promedios

```
media = df.apply(lambda x: x.mean())
print('Promedio python:', media.Python)
print('Promedio cython:', media.Cython)
```

```
print(f'Cython es {media.Python/media.Cython} veces mas rapido que Python')
```

### Barplot Python vs. Cython

```
sns.barplot(data=df)
plt.ylabel('Tiempo Segundos')
```

### Scatter

```
sns.scatterplot(data=df)
plt.ylabel('Tiempo Segundos')
```
